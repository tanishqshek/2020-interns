```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
inr = []
dates = []
gbp = []
latest_inr = []
latest_gbp = []
```

```python
#test
myurl = 'https://api.exchangeratesapi.io/2019-01-01'
latest_url = 'https://api.exchangeratesapi.io/latest'
response = requests.get(myurl)
response_for_latest = requests.get(latest_url)
r = response.json()
latest = response_for_latest.json()
print(latest['rates']['INR'])
```

```python
for i in range(1,32):
    if(i < 10):
        str1 = '0'+ str(i);
    else:
        str1 = str(i);
    try:
        myurl = 'https://api.exchangeratesapi.io/2019-01-'+str1
        response = requests.get(myurl)
        r = response.json()
        inr.append(r['rates']['INR'])
        dates.append(myurl[32:42])
        latest_inr.append(latest['rates']['INR'])
    except:
        print('Error')
        
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = dates
y = inr
# Specify the width and height of a figure in unit inches
fig = plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate', fontsize=12)
# Plot the data
plot1 = plt.plot(x,y, marker='o',label='INR')
plot2 = plt.plot(x,latest_inr, marker='^',label='Latest INR')
plt.legend()
plt.title('INR vs EUR with latest rates')
```

```python
for i in range(1,32):
    if(i < 10):
        str1 = '0'+ str(i);
    else:
        str1 = str(i);
    try:
        myurl = 'https://api.exchangeratesapi.io/2019-01-'+str1
        response = requests.get(myurl)
        r = response.json()
        gbp.append(r['rates']['GBP'])
        latest_gbp.append(latest['rates']['GBP'])
    except:
        print('Error')
```

```python
x = dates
y1 = inr
y2 = gbp
fig = plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate', fontsize=12)
# Plot the data
plot1 = plt.plot(x,y1, marker='o',label='INR')
plot2 = plt.plot(x,y2, marker='o',label='GBP')
plt.legend()
plt.title('INR and GBP vs EUR')
```
