```python
import pandas as pd
import json
```

```python
x = pd.read_json('data.json')
df = pd.DataFrame(x)
print(df['rates']['2019-01-02'])
```

```python
inr = []
dates = []
gbp = []
```

```python
str1 = ''

for i in range(1,32):
    if(i < 10):
        str1 = '0'+ str(i);
    else:
        str1 = str(i);
    try:    
        y = x['rates']['2019-01-'+str1]['INR']
        print(y)
        inr.append(y)
        dates.append('2019-01-'+str1)
    except:
        print('Error')
```

```python
print(dates)
```

```python
print(inr)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Create pandas dataframe
# df = pd.DataFrame(hist_data)
# Put dates on the x-axis
x = dates
# Put exchange rates on the y-axis
y = inr
# Specify the width and height of a figure in unit inches
fig = plt.figure(figsize=(15,6))
# Rotate the date ticks on the x-axis by degrees
plt.xticks(rotation=90)
# Set title on the
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate', fontsize=12)
# Plot the data
plot1 = plt.plot(x,y, marker='o',label='INR')
plt.legend()
```
