```python
import pandas as pd
import json
```

```python
x = pd.read_json('data.json')
latest = pd.read_json('latest-rates.json')
df = pd.DataFrame(x)
print(df['rates']['2019-01-02'])
```

```python
inr = []
dates = []
gbp = []
latest_inr = []
latest_gbp = []
```

```python
str1 = ''

for i in range(1,32):
    if(i < 10):
        str1 = '0'+ str(i);
    else:
        str1 = str(i);
    try:    
        y = x['rates']['2019-01-'+str1]['INR']
        print(y)
        inr.append(y)
        dates.append('2019-01-'+str1)
        latest_inr.append(latest['rates']['INR'])
    except:
        print('Error')
```

```python
print(latest_inr)
```

```python
print(dates)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = dates
y = inr
fig = plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate', fontsize=12)
# Plot the data
plot1 = plt.plot(x,y, marker='o',label='INR')
plot2 = plt.plot(x,latest_inr, marker='^',label='Latest INR')
plt.legend()
plt.title('INR vs EUR with latest rates')
```
