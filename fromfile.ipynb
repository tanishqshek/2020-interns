```python
import pandas as pd
import json
```

```python
x = pd.read_json('data.json')
latest = pd.read_json('latest-rates.json')
df = pd.DataFrame(x)
print(df['rates']['2019-01-02'])
```

```python
#making input currency dynamic
currency1_input = input("Enter your desired First Currency: ")
currency2_input = input("Enter your desired Second Currency: ")
```

```python
currency1 = []
dates = []
currency2 = []
latest_currency1 = []
latest_currency2 = []
```

```python
str1 = ''

for i in range(1,32):
    if(i < 10):
        str1 = '0'+ str(i);
    else:
        str1 = str(i);
    try:    
        y = x['rates']['2019-01-'+str1][currency1_input]
#         print(y)
        currency1.append(y)
        dates.append('2019-01-'+str1)
        latest_currency1.append(latest['rates'][currency1_input])
    except:
        print('Error')
```

```python
print(dates)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = dates
y = currency1
fig = plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate', fontsize=12)
# Plot the data
plot1 = plt.plot(x,y, marker='o',label=currency1_input)
plot2 = plt.plot(x,latest_currency1, marker='^',label='Latest '+ currency1_input)
plt.legend()
plt.title(currency1_input+ ' vs EUR with latest rates')
```
